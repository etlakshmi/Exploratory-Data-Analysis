```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(feature,target,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
re=LinearRegression()
```

```python
re.fit(xtrain,ytrain)
```

```python
r2score=re.score(xtrain,ytrain)
```

```python
print(r2score)
```

```python
re.coef_
```

```python
re.intercept_
```

```python
ypred=re.predict(xtest)
```

```python
print(ypred)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(ytest,ypred)
```

```python
print(r2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae=mean_absolute_error(ytest,ypred)
mse=mean_squared_error(ytest,ypred)
rmse=np.sqrt(mse)
r2=r2_score(ytest,ypred)
```

```python
print(f"MAE:{mae}\nMSE:{mse}\nRMSE:{rmse}\nAccuracy:{r2}")
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(feature,target,test_size=0.2,random_state=2)
```

```python
reg=LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
r2score=reg.score(xtrain,ytrain)
```

```python
print(r2score)
```

```python
ypred1=reg.predict(xtest)
```

```python
mae=mean_absolute_error(ytest,ypred1)
mse=mean_squared_error(ytest,ypred1)
rmse=np.sqrt(mse)
r2=r2_score(ytest,ypred1)
print(f"MAE:{mae}\nMSE:{mse}\nRMSE:{rmse}\nAccuracy:{r2}")
```

```python
train=re.score(xtrain,ytrain)
test=re.score(xtest,ytest)
print(f"Training Results:{train}")
print(f"Testing Results:{test}")
```

```python
from sklearn.linear_model import Ridge,Lasso
for i in range(50,100):
    l1=Lasso(alpha=i)
    l1.fit(xtrain,ytrain)
    train=l1.score(xtrain,ytrain)
    test=l1.score(xtest,ytest)
    print(f"{i} {train} {test}")
```

```python
for i in range(20,50):
    l2=Ridge(alpha=i)
    l2.fit(xtrain,ytrain)
    train=l2.score(xtrain,ytrain)
    test=l2.score(xtest,ytest)
    print(f"{i} {train} {test}")
```
